###### Import all Required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###### intutiation about Dataset

```python
df=pd.read_csv("globalterrorismdb_0718dist.csv",encoding="latin1")
```

```python
df.head()
```

```python
df.info()
```

```python
column_name=[]
for i in df.columns:
    column_name.append(i)
```

```python
column_name
```

```python
df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
```

```python
df.head()
```

```python
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Gang Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Killed']=df['Killed'].fillna(0)
df['Wounded']=df['Wounded'].fillna(0)
df['Total Casualty']=df['Killed']+df['Wounded']
```

```python
df.describe()
```

## Observations

- The dataset consists of terrorist attacks from the year 1970 to 2017.
- A maximum of 1,570 citizens were killed in an attack.
- A maximum of 8,191 citizens were wounded in an attack.
- The maximum number of injured citizens in an attack was 9,574.

## Yearwise Attack Analysis

```python
anual_attack=df['Year'].value_counts().sort_index().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
```

```python
anual_attack.head(10)
```

```python
anual_attack.plot(kind="bar",figsize=(15,5))
plt.title("Visual Analysis of Annual Attacks")
plt.xlabel("Years")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
print(anual_attack['Attacks'].min())
print(anual_attack['Attacks'].max())
```

- In 1971, there were the lowest number of attacks (471), according to the dataset.
- In 2014, there were the highest number of attacks ever (16,903), according to the dataset.

```python
annual_kill=df[["Year","Killed"]].groupby("Year").sum()
annual_kill.head()
```

```python
annual_wounded=df[["Year","Wounded"]].groupby("Year").sum()
annual_wounded.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

#Killed
annual_kill.plot(kind="bar",color="red",figsize=(15,13),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Year")
ax0.set_ylabel("Number of People Killed") 

#Wounded
annual_wounded.plot(kind="bar",color="orange",figsize=(15,13),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

- In 2014, terrorism was at its peak, with the maximum number of people killed and severe casualties.
- In 2015, it recorded the highest number of injured people among all the years.

### Regionwise Analysis

```python
df['Region'].unique()
```

```python
region_attack=pd.crosstab(df.Year,df.Region)
region_attack.head()
```

```python
region_attack.plot(kind="line",figsize=(15,10))
plt.title("Region wise attacks")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
region_attack_trans=region_attack.transpose()
```

```python
region_attack_trans.head()
```

```python
region_attack_trans['Total_casuality']=region_attack_trans.sum(axis=1)
```

```python
region_total_casuality=region_attack_trans['Total_casuality'].sort_values()
```

```python
region_total_casuality
```

```python
region_total_casuality.plot(kind="bar",fontsize=8)
```

#### Middle East & North africa faced a highest number of attacks while Australasia & Oceania faced lowest number of attacks in 1970-2017 

# CountryWise Attack Analysis

```python
country_attack=df["Country"].value_counts().head()
```

```python
country_attack
```

```python
country_attack.plot(kind="bar",fontsize=8)
plt.title("Country wise Attacks")
plt.xlabel("Countries")
# plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

```python
country_wise_casuality=df[["Country","Total Casualty"]].groupby("Country").sum().sort_values(by="Total Casualty",ascending=False)
country_wise_casuality.head(10)
```

```python
country_wise_casuality[:10].plot(kind="bar",fontsize=8)
plt.title("Country wise Casualties")
plt.xlabel("Countries")
plt.ylabel("Number of Casualties")
plt.show()
```

#### Iraq is the most affected countries followed by Pakistan, Afghanistan & India in terms of its casualities Countrywise.
